```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_df=pd.read_csv(r"D:\AI\Machine_Learning\supervised_learning\spam.csv")
```

```python
spam_df
```

```python
#inspect data
spam_df.groupby('Category').describe()
```

```python
#turning spam/ham into numerical data, create new column call spam
def label(s):  # or using lambda: spam_df['spam']=spam_df['Category'].apply(lambda x: 1 if x=='spam' else 0)
    if s=='spam':
        return 1
    else:
        return 0
spam_df['Spam_label']=spam_df['Category'].apply(label)
```

```python
X=spam_df['Message']
y=spam_df['Spam_label']
```

```python
#train/test split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
X_train
```

```python
#Find word count and store data as a matrix
cv=CountVectorizer()  #feature extraction on text data. Converts a collection text document into a matrix of token(word) counts [0;1]. That mean count number of each words that repeat in each document
X_train_count=cv.fit_transform(X_train.values)
```

```python
X_train_count.toarray()
```

```python
model=MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
#pre-test ham
email_ham=["hey wanna play footbal?"]
email_ham_count=cv.transform(email_ham)
model.predict(email_ham_count)
```

```python
#pre-test spam
email_spam=["reward money click"]
email_spam_count=cv.transform(email_spam)
model.predict(email_spam_count)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#test model
X_test_count=cv.transform(X_test)
y_pred=model.predict(X_test_count)
y_pred
```

```python
#evaluate
evaluate=accuracy_score(y_test,y_pred)
evaluate
```

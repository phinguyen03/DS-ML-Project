# Explaining


- Car_Name: Car_Name là tên hoặc mẫu xe cụ thể. Đó có thể là bất kỳ thương hiệu hoặc mẫu nào, chẳng hạn như "Honda Civic," "Toyota Corolla," hoặc "Ford Mustang."

- Year: Year chỉ ra năm sản xuất của chiếc xe. Nó biểu thị năm mà chiếc xe được sản xuất và thường được sử dụng để xác định tuổi của chiếc xe.

- Selling_Price: Selling_Price là giá bán mà chủ sở hữu hoặc đại lý định bán chiếc xe. Đây là số tiền mà người bán mong đợi nhận được khi trao đổi chiếc xe.

- Present_Price: Present_Price là giá thị trường hiện tại của chiếc xe. Nó đại diện cho giá trị ước tính của chiếc xe tại thời điểm bán, dựa trên tuổi tác, tình trạng, số dặm đã đi và nhu cầu trên thị trường.

- Kms_Driven: Kms_Driven chỉ ra tổng quãng đường hoặc số dặm đã đi của chiếc xe kể từ khi mua lần đầu hoặc từ khi chỉ số đo đạc đã được đặt lại. Nó cung cấp một ý tưởng về mức độ sử dụng rộng rãi của chiếc xe.

- Fuel_Type: Fuel_Type đề cập đến loại nhiên liệu mà chiếc xe sử dụng để di chuyển. Đó có thể là xăng/petrol, dầu diesel, điện, hybrid hoặc bất kỳ nguồn nhiên liệu khác.

- Seller_Type: Seller_Type mô tả loại người bán cung cấp chiếc xe để bán. Đó có thể là một đại lý hoặc một cá nhân. Đại lý thường bán xe chuyên nghiệp, trong khi cá nhân bán xe của riêng họ.

- Transmission: Transmission đề cập đến loại hộp số mà chiếc xe có. Nó xác định cách truyền động từ động cơ đến bánh xe. Các loại phổ biến bao gồm hộp số sàn (người lái thủ công chuyển số) và hộp số tự động (xe chuyển số tự động).

- Owner: Owner cho biết số lượng chủ sở hữu trước đó của chiếc xe. Nó cung cấp thông tin về lịch sử của chiếc xe và số người đã sở hữu và sử dụng nó trước người bán hiện tại.

```python
def rank_price(row):
    if row['Year'] >= 2014 and row['Present_Price'] >= 9.6 and row['Kms_Driven'] >= 2000 and row['Selling_Price'] > 9.0 :
        return 'high price'
    elif row['Year'] >= 2012 and row['Present_Price'] >= 6.0 and row['Kms_Driven'] >= 4000 and 5.0 <= row['Selling_Price'] <= 9.0 :
        return 'average price'
    elif row['Year'] >= 2010 and row['Present_Price'] >= 5.0 and row['Kms_Driven'] >= 4000 and 3.0 <= row['Selling_Price'] <= 4.9 :
        return 'low price'
    else:
      return 'super low price'

df['Rank Price'] = df.apply(rank_price, axis=1)
```

# Normalize

```python
y=df['Selling_Price']
```

```python
X=df.drop('Selling_Price',axis=1)
```

```python
X.info()
```

```python
category=X.select_dtypes(include='object').columns.tolist()
```

```python
numeric=X.select_dtypes(include=('float64','int64')).columns.tolist()
```

```python
print(f'Categories: {category}')
print(f'Numeric: {numeric}')
```

```python
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
```

```python
la=LabelEncoder()
for i in category:
  X[i]=la.fit_transform(X[i])
mi=MinMaxScaler()
for j in numeric:
  X[j]=la.fit_transform(X[j])
```

```python
X.head()
```

# Train

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear=LinearRegression(fit_intercept=True,copy_X=True,n_jobs=None,positive=False)
linear.fit(X_train, y_train)
```

```python
y_pre=linear.predict(X_test)
print(f'pre: {y_pre[:5]}')
print(f'y_true: {y_test[:5].tolist()}')
```

```python
from sklearn.metrics import  mean_squared_error, mean_absolute_error
```

```python
mse_linear = mean_squared_error(y_test, y_pre) #loss funct
mae_linear = mean_absolute_error(y_test, y_pre)
print(f'mearn squared: {mse_linear}')
print(f'mean absolute: {mae_linear }')
```

```python
li_score = linear.score(X_test, y_test)
print(li_score)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)
```

```python
X_train_poly,X_test_poly,y_train_poly,y_test_poly=train_test_split(X_poly,y,test_size=0.2, random_state=42)
```

```python
lin2=LinearRegression()
lin2.fit(X_train_poly,y_train_poly)
```

```python
y_pre2=lin2.predict(X_test_poly)
```

```python
mse2 = mean_squared_error(y_test_poly, y_pre2) #loss funct
mae2 = mean_absolute_error(y_test_poly, y_pre2)
print(f'mean squared: {mse2}')
print(f'mean absolute: {mae2 }')
```

```python
r2_score = lin2.score(X_test_poly, y_test)
print(r2_score)
```

# SVR model

```python
import numpy as np
```

```python
svr_pre=svr.predict(X_test)
```

# Randomforest model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
r=RandomForestRegressor(random_state=42,n_estimators=200,criterion='absolute_error'
,max_depth=10,min_samples_split=2
                        ,min_samples_leaf=1,min_weight_fraction_leaf=0.0
                        ,max_features='sqrt',min_impurity_decrease=0.0,n_jobs=2,verbose=0,ccp_alpha=0.0,bootstrap=True,oob_score=False).fit(X_train,y_train)
```

```python
pre_ran=r.predict(X_test)
```

```python
print(f'predict: {pre_ran[:5]}')
print(f'y: {y_test[:5].tolist()}')
```

```python
mse4 = mean_squared_error(y_test, pre_ran) #loss funct
mae4 = mean_absolute_error(y_test,pre_ran)
print(f'mean squared error: {mse4}')
print(f'mean absolute error: {mae4}')
```

```python
ran_score=r.score(X_test,y_test)
ran_score
```
